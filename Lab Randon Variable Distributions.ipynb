```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")
import statistics as stats
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
numericals = data.select_dtypes(np.number)
```

```python
numericals.head()
```

```python
for i in numericals:
    sns.displot(numericals[i])
    plt.show()

```

```python
m = numericals["Total Claim Amount"].mean()
s = numericals["Total Claim Amount"].std()
sns.distplot(numericals["Total Claim Amount"])
sns.distplot(stats.norm(m,s).rvs(1000))
```

```python
transformed = np.log(numericals["Total Claim Amount"])
mt = transformed.mean()
st = transformed.std()
```

```python
sns.distplot(transformed)
sns.distplot(stats.norm(mt,st).rvs(1000))
```
